```python
import pandas as pd

def parse_score(score):
    try:
        home_score, away_score = map(int, score.split('–'))
        return home_score, away_score
    except:
        return None, None

# Aplicar la función de parsing a la columna 'Score'
laliga1819[['HomeScore', 'AwayScore']] = laliga1819['Score'].apply(lambda x: pd.Series(parse_score(x)))

# Filtrar partidos en los que el Girona fue el equipo local y ganó por 2 o más goles
home_wins = laliga1819[(laliga1819['Local'] == 'Girona') & (laliga1819['HomeScore'] - laliga1819['AwayScore'] >= 2)]

# Filtrar partidos en los que el Girona fue el equipo visitante y ganó por 2 o más goles
away_wins = laliga1819[(laliga1819['Visitante'] == 'Girona') & (laliga1819['AwayScore'] - laliga1819['HomeScore'] >= 2)]

# Contar la cantidad de partidos en los que el Girona ganó por 2 o más goles
total_wins = len(home_wins) + len(away_wins)

print(f'Cantidad de partidos en que el Girona ganó por 2 o más goles en la temporada 18/19: {total_wins}')
```

```python
import pandas as pd

# Suponiendo que df_all ya está cargado como DataFrame


# Filtrar los partidos de la temporada 21/22
start_date = '2021-08-01'
end_date = '2022-05-31'
df_2122 = df_all[(df_all['Date'] >= start_date) & (df_all['Date'] <= end_date)]

# Paso 2: Filtrar para Alcorcón como local o visitante
df_alcorcon = df_2122[(df_2122['Local'] == 'Alcorcón') | (df_2122['Visitante'] == 'Alcorcón')]

# Paso 3: Calcular la diferencia de goles y contar los partidos perdidos por al menos 2 goles
def calculate_goal_difference(row):
    local_score = row['HomeScore']
    away_score = row['AwayScore']
    
    if row['Local'] == 'Alcorcón':
        return away_score - local_score
    else:
        return local_score - away_score

# Aplicar la función para obtener la diferencia de goles
df_alcorcon['GoalDifference'] = df_alcorcon.apply(calculate_goal_difference, axis=1)

# Contar partidos que Alcorcón perdió por al menos 2 goles de diferencia
lost_by_2_or_more = df_alcorcon[df_alcorcon['GoalDifference'] >= 2].shape[0]

print(f'Cantidad de partidos que el Alcorcón perdió por al menos 2 goles de diferencia en la temporada 21/22: {lost_by_2_or_more}')
```

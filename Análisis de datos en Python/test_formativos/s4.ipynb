```python
import pandas as pd

# Supongamos que ya tienes el DataFrame 'df'

# Contar el número de días que el cajero puede ser reabastecido
df['Días de Reabastecimiento'] = df[['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']].sum(axis=1)

# Aplicar los filtros:
# 1. Reabastecimiento al menos 4 días a la semana
# 2. Al menos uno de los días en fin de semana (Sábado o Domingo)
# 3. No tiene turno de noche
cajeros_filtrados = df[
    (df['Días de Reabastecimiento'] >= 4) &
    ((df['Sábado'] == 1) | (df['Domingo'] == 1)) &
    (df['Turno noche'] == 0)
]

# Contar el número de cajeros que cumplen con todos los criterios
numero_cajeros_filtrados = cajeros_filtrados.shape[0]

print(f"Número de cajeros que pueden ser reabastecidos al menos 4 días a la semana, con al menos un día en fin de semana y sin turno de noche: {numero_cajeros_filtrados}")
```

```python
laliga2122 = pd.read_excel('LaLiga21 22.xlsx', )
laliga2021 = pd.read_excel('LaLiga20 21.xlsx', )
laliga1920 = pd.read_excel('LaLiga19 20.xlsx', )
laliga1819 = pd.read_excel('LaLiga18 19.xlsx', )
laliga1718 = pd.read_excel('LaLiga17 18.xlsx', )
laliga1617 = pd.read_excel('LaLiga16 17.xlsx', )
seg2122 =  pd.read_excel('LaLiga123 21 22.xlsx', )
seg2021 =  pd.read_excel('LaLiga123 20 21.xlsx', )
seg1920 =  pd.read_excel('LaLiga123 19 20.xlsx', )
seg1819 =  pd.read_excel('LaLiga123 18 19.xlsx', )
seg1718 =  pd.read_excel('LaLiga123 17 18.xlsx', )
seg1617 =  pd.read_excel('LaLiga123 16 17.xlsx', )
```

```python

# Combinar los DataFrames
df_all = pd.concat([laliga, segunda])

```

```python
import pandas as pd

# Filtrar el DataFrame para las posiciones del Betis y del Sevilla
betis = df_all[df_all['Local'] == 'Betis']
sevilla = df_all[df_all['Local'] == 'Sevilla']

# Asegúrate de que los datos estén alineados por fecha
df = pd.merge(betis[['Date', 'Score']], sevilla[['Date', 'Score']], on='Date', suffixes=('_betis', '_sevilla'))

# Contar cruces
crossings = 0
previous_betis = df['Score_betis'].iloc[0]
previous_sevilla = df['Score_sevilla'].iloc[0]

for i in range(1, len(df)):
    current_betis = df['Score_betis'].iloc[i]
    current_sevilla = df['Score_sevilla'].iloc[i]
    
    if (previous_betis < previous_sevilla and current_betis > current_sevilla) or \
       (previous_betis > previous_sevilla and current_betis < current_sevilla):
        crossings += 1
    
    previous_betis = current_betis
    previous_sevilla = current_sevilla

print(f'Cantidad de cruces: {crossings}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que ya tienes df_all cargado con tus datos
# Filtrar los datos para Betis y Sevilla
betis = df_all[df_all['Local'] == 'Betis']
sevilla = df_all[df_all['Local'] == 'Sevilla']

# Asegúrate de que los datos estén ordenados por fecha
betis = betis.sort_values(by='Date')
sevilla = sevilla.sort_values(by='Date')

# Extraer fechas y posiciones (en este caso, 'Score' se usa como proxy para la posición)
fechas_betis = betis['Date']
posiciones_betis = betis['Score']

fechas_sevilla = sevilla['Date']
posiciones_sevilla = sevilla['Score']

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar las posiciones del Betis
ax.plot(fechas_betis, posiciones_betis, color='green', linestyle='dashed', linewidth=3, marker='o', markerfacecolor='blue', markersize=8, label='Betis')

# Graficar las posiciones del Sevilla
ax.plot(fechas_sevilla, posiciones_sevilla, color='red', linestyle='dashed', linewidth=3, marker='o', markerfacecolor='orange', markersize=8, label='Sevilla')

# Añadir etiquetas y leyenda
ax.set_xlabel('Fecha')
ax.set_ylabel('Posición')
ax.set_title('Evolución de la posición del Betis y del Sevilla')
ax.legend()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import numpy as np
import glob
import plotly.express as px
import matplotlib.pyplot as plt
plt.style.use('default')
import warnings
warnings.filterwarnings('ignore')
```

```python
data = {
    'Ensayo': ['Ensayo 1', 'Ensayo 2', 'Ensayo 3', 'Ensayo 4', 'Ensayo 5', 'Ensayo 6', 'Ensayo 7'],
    'Buenas': [7, 6, 9, 7, 11, 10, 11],
    'Malas': [6, 3, 1, 7, 1, 3, 4],
    'Omitidas': [2, 6, 5, 1, 4, 2, 0],
    'Total': [15, 15, 15, 15, 15, 15, 15]
}
df = pd.DataFrame(data)
statistics = df.describe()

print("Estad√≠sticas:")
print(statistics)

plt.figure(figsize=(10, 6))
plt.plot(df['Ensayo'], df['Buenas'], marker='o', label='Buenas')
plt.plot(df['Ensayo'], df['Malas'], marker='o', label='Malas')
plt.plot(df['Ensayo'], df['Omitidas'], marker='o', label='Omitidas')
plt.title('Desempe√±o de Evaluaci√≥n del Estudiante')
plt.xlabel('Ensayo')
plt.ylabel('Cantidad')
plt.legend()
plt.grid(True)
plt.show()
plt.figure(figsize=(20, 12))
df[['Buenas', 'Malas', 'Omitidas']].boxplot()
plt.title('Distribuci√≥n de Resultados por Categor√≠a')
plt.ylabel('Cantidad')
plt.grid(True)
plt.show()
```

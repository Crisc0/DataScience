```python
#código de inicio
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

#### 2. Extraccion de la lista de anuncios de trabajo arrojados por Linkedin.

```python
#complete este código
position = 'data scientist'
location = 'chile'
url_search = 'https://www.linkedin.com/jobs/search/?keywords=%s&location=%s' % (position, location)
```

```python
#código de inicio

#Para evitar que la página web piense que usted es un bot, al realizar el request utilice algunos de los siguientes encabezados: 
head = {'User-Agent': 'Mozilla/5.0'}
#head = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.84 Safari/537.36'}
#head = {'user-agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Mobile Safari/537.36'}
#head = {'user-agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Mobile Safari/537.36'}
#head = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.84 Safari/537.36',}
```

```python
response = requests.get(url_search, headers=head)
soup = BeautifulSoup(response.content, 'html.parser')
joblist = soup.find('ul', class_="jobs-search__results-list")

```

```python
joblist
```

#### 3. Seleccionando solo el primer anuncio de la lista, y extrayendo la información de:  título del trabajo, nombre de la compañía, localización, y URL del anuncio
Nota: Por localización se entiende la ciudad, comuna o municipio indicado en el anuncio.

```python
primer_trabajo = joblist.find('li') 
title = primer_trabajo.find('h3', class_='base-search-card__title').get_text(strip=True)
company = primer_trabajo.find('h4', class_='base-search-card__subtitle').get_text(strip=True)
location = primer_trabajo.find('span', class_='job-search-card__location').get_text(strip=True)
joburl = primer_trabajo.find('a', class_='base-card__full-link')['href']

print(f'title: {title}')
print(f'company: {company}')
print(f'location: {location}')
print(f'joburl: {joburl}')
```

#### 4. En base a los puntos anteriores, programe una rutina para extraer la información de localización,  título del trabajo, nombre de la compañía, localización, y URL del anuncio para todos los trabajos arrojados por su búsqueda de Linkedin, almacenando los datos en un dataframe de pandas.

```python
jobs = joblist.find_all('li')
job_data = []
for job in jobs:
    Title = job.find('h3', class_='base-search-card__title').get_text(strip=True)
    Company = job.find('h4', class_='base-search-card__subtitle').get_text(strip=True)
    Location = job.find('span', class_='job-search-card__location').get_text(strip=True)
    Url = primer_trabajo.find('a', class_='base-card__full-link')['href']
    job_data.append({
        'Title': Title,
        'Company': Company,
        'Location': Location,
        'Url ': Url
    })
df_jobs = pd.DataFrame(job_data)
print(df_jobs.head(5))
```
